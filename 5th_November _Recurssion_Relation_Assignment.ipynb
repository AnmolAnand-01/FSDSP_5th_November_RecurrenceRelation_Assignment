# 2. Given a recurrence relation, solve it using the substitution method:

# a. T(n) = T(n-1) + c

# b. T(n) = 2T(n/2) + n

# c. T(n) = 2T(n/2) + c

# d. T(n) = T(n/2) + c

**Solution:**

**a. T(n) = T(n-1) + c**

Assume Base case at n=1, 

Therefore T(n) = T(n-1) + c if n > 1

and T(n) = c if n == 1

=> T(n) = T(n-1) + c  -- eq 1

Substitute the value of T(n-1) in eq 1

=> T(n) = T(n-2) + 2c  -- eq 2

Substitute the value of T(n-2) in eq 2

=> T(n) = T(n-3) + 3c  -- eq 3

Perform the substitution for "k" times

=> T(n) = T(n-k) + k*c  -- eq 4

using the base case T(n) = c for n == 1

=> n - k = 1, therefore k = n - 1

Substitute k = n - 1 in eq 4

=> T(n) = T(n-(n-1)) + (n-1)*c

=> T(n) = T(1) + (n-1)*c

=> T(n) = c + (n-1)*c 

=> T(n) = n*c

therefore T(n) = O(n)

**b. T(n) = 2T(n/2) + n**

Assume Base case at n=1,

Therefore T(n) = 2T(n/2) + n if n > 1

and T(n) = c if n == 1

=> T(n) = 2T(n/2) + n -- eq 1

Substitute the value of T(n/2) in eq 1

=> T(n) = (2^2)T(n/(2^2)) + 2n -- eq 2

Substitute the value of T(n/(2^2)) in eq 1

=> T(n) = (2^3)T(n/(2^3)) + 3n -- eq 3

Perform the substitution for "k" times

=> T(n) = (2^k)T(n/(2^k)) + k*n -- eq 4

using the base case T(n) = c for n == 1

=> n/(2^k) = 1, therefore k = log2(n)

Substitute k = log2(n) in eq 4

=> T(n) = (2^(log2(n)))T(n/(2^(log2(n)))) + (log2(n))*n 

=> T(n) = n*T(1) + n*(log2(n))

=> T(n) = n*c + n*(log2(n))

Therefor T(n) = O(n(log2(n)))

**c. T(n) = 2T(n/2) + c**

Assume Base case at n=1,

Therefore T(n) = 2T(n/2) + c if n > 1

and T(n) = c if n == 1

=> T(n) = 2T(n/2) + c -- eq 1

Substitute the value of T(n/2) in eq 1

=> T(n) = (2^2)T(n/(2^2)) + 2c -- eq 2

Substitute the value of T(n/(2^2)) in eq 1

=> T(n) = (2^3)T(n/(2^3)) + 3c -- eq 3

Perform the substitution for "k" times

=> T(n) = (2^k)T(n/(2^k)) + k*c -- eq 4

using the base case T(n) = c for n == 1

=> n/(2^k) = 1, therefore k = log2(n)

Substitute k = log2(n) in eq 4

=> T(n) = (2^(log2(n)))T(n/(2^(log2(n)))) + (log2(n))*c

=> T(n) = n*T(1) + c*(log2(n))

=> T(n) = n*c + c*(log2(n))

=> T(n) = c(n + log2(n))

Therefor T(n) = O(n)

**d. T(n) = T(n/2) + c**

Assume Base case at n=1,

Therefore T(n) = T(n/2) + c if n > 1

and T(n) = c if n == 1

=> T(n) = T(n/2) + c -- eq 1

Substitute the value of T(n/2) in eq 1

=> T(n) = T(n/(2^2)) + 2c -- eq 2

Substitute the value of T(n/(2^2)) in eq 1

=> T(n) = T(n/(2^3)) + 3c -- eq 3

Perform the substitution for "k" times

=> T(n) = T(n/(2^k)) + k*c -- eq 4

using the base case T(n) = c for n == 1

=> n/(2^k) = 1, therefore k = log2(n)

Substitute k = log2(n) in eq 4

=> T(n) = T(n/(2^(log2(n))) + (log2(n))*c

=> T(n) = T(1) + c(log2(n))

=> T(n) = c + c(log2(n))

Therefor T(n) = O(log2(n))
